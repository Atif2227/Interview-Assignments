# Data Manupualtion For Question 1.

#### Let's check the pattern of all the features when the data is sorted by 'Revenue' and grouped by 'App'.

```python
# Apps with with sum of all the numerical features in ascending order.

gg_df.groupby('App').sum('Revenue').sort_values(by='Revenue', ascending=True)
```

From the Above table it is clear that, there are many Apps with zero revenue and all of them share a common pattern. They have few to few thousands Ad Request, but there was no Ad Responses and hence no Fill Rate, Impressions, Render rate, 1St Clicks and 2nd Clicks respectively.

```python
# Apps with with sum of all the numerical features in descending order.

gg_df.groupby('App').sum('Revenue').sort_values(by='Revenue', ascending=False)
```

From the above table it is clear that the Apps with top Revenue have very large number of Ad requests with large number of Ad Responses. When Ad Response is large other features such as Fill Rate, Impressions, Render Rate, 1st Click and 2nd Click are also in large numbers.

## Question 1st
### How would you increase the overall revenue for an app?

## Answer:
### Findings from the above tables - 
1. Top Apps with highest sum of Revenue has high Ad Requests which corresponds to higher values of other features as well.
2. Top apps with lowest sum of Revenue has either very less or zero Ad Requests with zero values for other features.
3. It is clear that the process of revenue generation for an app is a wise sequential method.
4. The sequence goes like this - 
    a) Make high Ad Request which would correspond to a high Ad Response.
    b) High Ad Response would generate high Impression which would correspond to a high Fill Rate and high Render Rate.
    c) When all these parameters are set high there would be high 1st Click and high 2nd Click.
    
5. Thus for increasing the over all revenue of an App, there must be high Ad Request first and then other feature values should also be high.
